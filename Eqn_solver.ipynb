```
from colored import fg,bg,attr
import time

def Linear(a,b):
  solve_linear= -b/a
  print('x = '+str(solve_linear))
def Quadratic(a,b,c):
  d = ((b)**2 - 4*a*c)
  if ((b)**2 - 4*a*c) > 0:
    x1 = (-b + d**0.5)/2*a
    x2 = (-b - d**0.5)/2*a
    print('x = '+str(x1)+' and '+str(x2))
  elif (b**2 - 4*a*c) < 0:
    x1 = -b/2*a
    x2 = (abs(d)**0.5)/2*a
    print('x = '+str(x1)+' + '+str(x2)+'j and '+str(x1)+' - '+str(x2)+'j')
def Cubic(a,b,c,d):
  lst=[]
  for i in range(-10000,10000):
    if ((a*((i)**3) + b*((i)**2) + c*(i) + d)==0):
      lst.append(i)
  print('x = '+str(lst))
print('%s%s Want to solve algebric equations ? %s'% (fg(1),attr('bold'),attr('reset')))
ask=input()
time.sleep(0.5)
reply_Y=['Yes','yes','Y','y','YES']
time.sleep(0.1)
if ask in reply_Y:
  ask_type=input('Linear or quadratic or cubic:\n')
  time.sleep(1)
  if ask_type=='Linear' or ask_type=='linear':
    print('%s%s Standard Linear form :: Ax + B = 0 \nSo comparing to above equation: write values of a and b -_-\n\n %s'%(fg(10),attr('bold'),attr('reset')))
    a=float(input('a: '))
    b=float(input('b: '))
    Linear(a,b)
  elif ask_type=='Quadratic' or ask_type=='quadratic':           
    print('%s%s Standard quadratic form :: Ax^2 + Bx + C = 0 \nSo comparing to above equation: write values of a, b and c -_-\n\n %s'%(fg(10),attr('bold'),attr('reset')))
    a = float(input('a: '))
    b = float(input('b: '))
    c = float(input('c: '))
    Quadratic(a,b,c)
  elif ask_type=='Cubic' or ask_type=='cubic':
    print('%s%s Standard quadratic form :: Ax^3 + Bx^2 + Cx + D = 0 \nSo comparing to above equation: write values of a, b, c ans d -_-\n\n %s'%(fg(10),attr('bold'),attr('reset')))
    a=float(input('a: '))
    b=float(input('b: '))
    c=float(input('c: '))
    d=float(input('d: '))
    Cubic(a,b,c,d)
```
